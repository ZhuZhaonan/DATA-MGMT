## 4. Raw data

```
# 2016 basic information
import pandas as pd
df1= pd.read_csv("https://raw.githubusercontent.com/ma-ji/RICF/master/2016/2016_basic.tsv",sep='\t')
df_bas=df1[["ricf_oid","ba_cn","ba_nfe","ba_gvm","ba_pgv","ba_ntr","ba_fdt","ba_bn","ba_evi"]] 
df_bas.rename(columns={'ba_evi':'ba_evl'}, inplace = True)
print(df_bas.head(5))
```

```
#2016 Financial activities

df2= pd.read_csv('https://raw.githubusercontent.com/ma-ji/RICF/master/2016/2016_financial_activities.tsv',sep='\t')
df_act=df2[['ricf_oid','fa_dine_this','fa_dinu_this']]
print(df_act.head(5))
```

```
# 2016 cash flow

df3= pd.read_csv('https://raw.githubusercontent.com/ma-ji/RICF/master/2016/2016_cash_flow.tsv',sep='\t')
df_cash= df3[['ricf_oid','cf_govc']]
print(df_cash.head(5))
```

```
# 2016 Financial position

df4= pd.read_csv('https://raw.githubusercontent.com/ma-ji/RICF/master/2016/2016_financial_position.tsv',sep='\t')
list(df4)
df_pos= df4[['ricf_oid','fp_tt_end']]
print(df_pos.head(5))
```

```
# Wang_2018
df= pd.read_excel('Wang_2018_revised.xlsx')
df.info()
```

```
df_wang=df[['ricf_oid','Government-Link','Scope','Province']]
print(df_wang.head(5))
```

```
# meger df_bas and df_act
df5=df_bas.merge(df_act, on="ricf_oid", how="inner")
print(df5.head(5))

# meger df_5 and df_cash
df6= df5.merge(df_cash,  on="ricf_oid", how="inner")
print(df6.head(5))

# meger df6 and df_pos
df7=df6.merge(df_pos,  on="ricf_oid", how="inner")
print(df7.head(5))

# meger Wang_2018
data_raw= df7.merge(df_wang,  on="ricf_oid", how="inner")
print(data_raw.head(5))
```

```
# save raw data
data_raw.to_csv('data_raw.csv')
```

## 5. Cleaning data

```
data_raw.shape
data_raw.dtypes
data_raw.info()
```

```
print(data_raw.head(5))
```

```
# coding ba_ntr

data_raw1=data_raw.replace('公募基金会',1)
data_raw2=data_raw1.replace('非公募基金会',0)
data_raw3=data_raw2
data_raw3
```

```
# coding governmet-link
data_raw4=data_raw3.replace('GLF',1)
data_raw5=data_raw4.replace('NGF',0)
data_raw5
```

```
# coding scope
data_raw6=data_raw5.replace('National',1)
data_raw7=data_raw6.replace('Regional',0)
data_raw7
```

```
# coding evalution result
data_raw7.ba_evl.fillna('None').str.contains('A')
df_e=data_raw7[data_raw5.ba_evl.fillna('None').str.contains('A')]
df_e['ba_evl']=[s.strip('A') for s in df_e.ba_evl]
df_e.info()
```

```
# cleaning ba_fdt count ba_age
df_e['ba_age']=df_e['ba_fdt'].str[0:4]
df_e=df_e[ ~ df_e['ba_age'].str.contains('发起人') ]
df_e['ba_age']=df_e['ba_age'].astype('float')
df_e['ba_age']=2015-df_e['ba_age']
df_e
```

```
# donation income of 2016
df_e['fa_dine_this']=df_e['fa_dine_this'].apply(lambda x: x.replace(',', '')).astype('float')
df_e['fa_dinu_this']=df_e['fa_dinu_this'].apply(lambda x: x.replace(',', '')).astype('float')
df_e["fa_dint"]=df_e["fa_dine_this"]+df_e["fa_dinu_this"]
df_e
```

```
# cleaning fp_tt_end
df_e=df_e[ ~ df_e['fp_tt_end'].str.contains('单位：人民币元') ]
df_e['fp_tt_end']=df_e['fp_tt_end'].apply(lambda x: x.replace(',', '')).astype('float')
df_e
```

```
# cleaning  cf_govc
df_e['cf_govc']=df_e['cf_govc'].apply(lambda x: x.replace(',', '')).astype('float')
df_e
```

```
df_e.info()
```

```
# count ba_gvof
df_e["ba_gvof"]=df_e["ba_gvm"]+df_e["ba_pgv"]
df_e.rename(columns={'Government-Link':'sc_gongo'}, inplace = True)
df_e.rename(columns={'fp_tt_end':'fp_tt'}, inplace = True)
df_e.rename(columns={'Scope':'sco_na'}, inplace = True)
```

```
# 单位：万元
df_e['fa_dint']=df_e['fa_dint']/10000
df_e['fp_tt']=df_e['fp_tt']/10000
df_e['cf_govc']=df_e['cf_govc']/10000
df_e
```

```
# 设置小数点位数
import numpy as np
df_e['fa_dint']=df_e['fa_dint'].round(decimals=2)
df_e['fp_tt']=df_e['fp_tt'].round(decimals=2)
df_e['cf_govc']=df_e['cf_govc'].round(decimals=2)
```

```
df_e
```

```
df_e = df_e.dropna(subset=['ba_ntr'])
df_e = df_e.dropna(subset=['ba_bn'])
```

```
df_e.info()
```

```
data_clean=df_e
data_clean.to_csv('data_clean.csv')
data_clean
```
